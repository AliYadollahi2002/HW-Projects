## 1 Slotted-ALOHA Capacity

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

### 1.1 Slotted ALOHA Simulation

```python
def simulate_aloha(n_users , n_timeslots , prob):
    data = [1 , 0]
    probability = [prob , 1-prob] 
    out = 0
    for slot in range(n_timeslots):
        s = 0
        for n in range(n_users):
            r = random.choices(data, probability)
            s = s + r[0]
        if s == 1:
            out = out + 1
    return out/n_timeslots
```

### 1.2 Verifying Slotted ALOHA Capacity

```python
prob = np.arange(0 , 1 , 0.001)
rate = []
for p in prob:
    rate.append(simulate_aloha(10 , 1000 , p))
    #print(simulate_aloha(10 , 1000 , p)
    
plt.plot(prob , rate)
```

we can see that the curve above matches the theoretical curve.

## 3 GPS signal

### 3.1

```python
random.seed(400102233)
sequence = []
for i in range(16):
    r = random.choices([1 , -1] , [0.5 , 0.5])
    sequence.append(r[0])
    
print(sequence)
print(len(sequence))
```

### 3.2

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sequence of bits
bits = sequence  # Replace with your own sequence of bits

# Time parameters
fs = 1000  # Sample rate
T = len(bits)  # Duration in seconds (equal to the number of bits)
t = np.arange(0, T, 1/fs)  # Time vector

# Generate the pulse wave from the bit sequence
pulse_wave = np.repeat(bits, int(fs))
bpsk = np.sin(12* np.pi * t)*pulse_wave

# Plot the pulse wave
plt.figure(figsize=(10,5))
plt.plot(t, pulse_wave , color = 'red')
plt.xlabel('Time (s)')
plt.title('CDMA BPSK')
plt.show()
plt.figure(figsize=(10,5))
plt.plot(t, bpsk , color = 'blue')
plt.xlabel('Time (s)')
plt.title('CDMA BPSK')
#plt.xlim(0 , 2)
plt.show()
plt.figure(figsize=(10,5))
plt.plot(t, pulse_wave , color = 'red' , linewidth = 2)
plt.plot(t, bpsk , color = 'blue')
plt.xlabel('Time (s)')
plt.title('CDMA BPSK')
plt.show()
```
